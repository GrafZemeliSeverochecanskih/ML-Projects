Let us see the distribution of all variable without missing values with except for Cabin variable, because we will need to make a preprocessing before.

```python
columns = list(train_validation_df.columns.copy())
columns.remove("Cabin")
```

```python
def is_discrete(data: pd.core.frame.DataFrame, column: str) -> bool:
    return True if data[column].dtype == "object" or data[column].nunique() < 20 else False
```

```python
for column in columns:
    plt.figure(figsize=(10, 5))
    if is_discrete(train_validation_df, column):
        sns.histplot(data = train_validation_df[column].copy().dropna(), discrete=True)
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
    else:
        sns.histplot(train_validation_df[column].copy().dropna(), kde=True)
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
    plt.show()
```

We can that all numerical variables with except for Age are extremely positevily skewed. For cabin variable, let its value be equal to the first letter to group the cabins in some way

## Data preprocessing

```python
train_validation_df["Cabin"] = train_validation_df["Cabin"].str[0]
```

```python
def print_missing_values(data: pd.core.frame.DataFrame) -> None:
    for i in data.columns:
        print(i, (data[i].isnull().sum()/len(data[i]))*100)
    return None
```

```python
print_missing_values(train_validation_df)
```

Because the amount of missing variables is less than 3% for all features, the missing values can be easily replaced. In case of categorical variables, we will replace them by feature mode. In case of numerical variables, we will replace them by feature median.

```python
def fill_null_values(data: pd.core.frame.DataFrame) -> None:
    for column in columns:
        if is_discrete(data, column):
            most_frequent_value = data[column].mode().iloc[0]
            data[column] = data[column].fillna(most_frequent_value)
        else:
            data[column] = data[column].fillna(data[column].median())
    most_frequent_value = data["Cabin"].mode().iloc[0]
    data["Cabin"] = data["Cabin"].fillna(most_frequent_value)
```

```python
fill_null_values(train_validation_df)
```

Check if there are still missing values

```python
print_missing_values(train_validation_df)
```

As our numerical data are extremely skewed, let us use IQR to deal with the outliers. We will make all values below or upper the 1.5 * IQR equalt to boundary value.

```python
def prepare_numerical_features(data: pd.core.frame.DataFrame):
    for i in ["VRDeck", "Spa", "ShoppingMall", "FoodCourt", "RoomService"]:
        q3, q1 = np.percentile(data[i], [75 ,25])
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        data[i] = np.where(data[i] < lower_bound, lower_bound, data[i])
        data[i] = np.where(data[i] > upper_bound, upper_bound, data[i])
        data[i] = np.log10(data[i]+1)
```

Check the uodated distributions

```python
prepare_numerical_features(train_validation_df)
for column in columns:
    plt.figure(figsize=(10, 5))
    if not is_discrete(train_validation_df, column):
        sns.histplot(train_validation_df[column].copy().dropna(), kde=True)
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
    plt.show()
```

Let us analyze correlation matrix to detect the multicollinear features.  

```python
sns.heatmap(data=train_validation_df[["VRDeck", "Spa", "ShoppingMall", "FoodCourt", "RoomService", "Age"]].corr())
```

No multicollinear features were detected. So, we can move to encoding categorical variables via LabelEncoder provided by scikit-learn library. (https://github.com/scikit-learn/scikit-learn/blob/b7d45ad56/sklearn/preprocessing/_label.py#L30)

```python
def encode_categorical_variable(data: pd.core.frame.DataFrame):
    le = LabelEncoder()
    for i in columns:
        if data[i].dtype not in ["float64", "int64"]:
            data[i] = le.fit_transform(train_validation_df[i])
    data["Cabin"] = le.fit_transform(data["Cabin"])
```

```python
encode_categorical_variable(train_validation_df)
```

## Model selection

Let us prepare X matrix and y vector for learning process and separate the dataframe in training and validation datasets.

```python
X, y = train_validation_df.copy().drop("Transported", axis = 1), train_validation_df["Transported"].values
X = X.values
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size = 0.25)
```

```python
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
y_pred = logReg.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
mlpc = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
mlpc.fit(X_train, y_train)
y_pred = mlpc.predict(X_validation)
accuracy = accuracy_score(y_validation, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_validation, y_pred))
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'max_depth': 3,
    'eta': 0.1,
    'seed': 42
}
num_rounds = 100
bst = xgb.train(params, dtrain, num_rounds)
dvalidation = xgb.DMatrix(X_validation)
y_pred_prob = bst.predict(dvalidation)
y_pred = [1 if prob > 0.5 else 0 for prob in y_pred_prob]
accuracy = accuracy_score(y_validation, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_validation, y_pred))
```
